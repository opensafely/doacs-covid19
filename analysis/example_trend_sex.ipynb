```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../output/measure_doacs_by_sex.csv", usecols=["date", "sex", "on_doac", "population"])
df["date"] = pd.to_datetime(df["date"])
female_subset=df.loc[df["sex"] == "F", :]
male_subset=df.loc[df["sex"] == "M", :]
x1_coordinates=female_subset["date"].dt.strftime('%B-%Y')
x2_coordinates=male_subset["date"].dt.strftime('%B-%Y')
y1_coordinates=female_subset["on_doac"]
y2_coordinates=male_subset["on_doac"]
plt.plot(x1_coordinates, y1_coordinates)
plt.plot(x2_coordinates, y2_coordinates)
```
